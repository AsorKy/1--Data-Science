##3)-Object Oriented Methods

One posible customization could be the posibility of having multiple plots whith different scales but being the plots within the same figure.

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig, ax1 = plt.subplots()
ax1.plot(x,y,'blue')
ax2 = ax1.twinx()
ax2.plot(y,x,'green')

ax1.set_xlabel('x')
ax2.set_xlabel('x')
ax1.set_ylabel('x^2')
ax2.set_ylabel('x^-2')

fig.set_size_inches(5,3)
fig.set_dpi(100)
```

We can create sub plots, or more specifically, "subaxes" such that one of the plots be within the other with a reduced size. This can be helpfull to make comparisons.

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.6,0.5])
axes_2 = fig.add_axes([0.145,0.36,0.3,0.2])
axes.plot(x,y,'b')
axes_2.plot(y,x,'r--')
axes_2.set_facecolor('white')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(x,y*y,'r')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b*-')
ax2.plot(x,y,'g*-')
plt.show()
```

```python
x = np.linspace(0,10,100)

fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(5,3))
axes[0,0].plot(x,np.sin(x),'b*-')
axes[0,1].plot(x,np.cos(x),'g*-')
axes[1,0].plot(x,np.tan(x),'y*-')
axes[1,1].plot(x,1/np.sin(x),'r*-')

fig.tight_layout()
plt.show()
```

##4)-Leyends, Labels, Titles and size

```python
x = np.linspace(1,10,100)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize = (8,5))
axes[0].plot(x,y, label='Sin(x)')
axes[1].plot(x,np.cos(x),'r:',label='Cos(x)')

axes[0].set_title('Sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].set_title('Cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')

axes[0].legend()
axes[1].legend()

fig.tight_layout()
plt.show()
```

We can be more specific and specify the RGB color properties, the line width and the line style.

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+1,color = '#D426C8', linestyle='-') # Color RGB
ax.plot(x,x+2,color = '#66FF89',linestyle='--')

ax.plot(x,x+3,color = 'blue', alpha=0.8, linewidth=5,linestyle='dashed') #Color comn
ax.plot(x,x+4, color = 'black',linestyle=':')
```

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.plot(x,x+3,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "#37D842")
ax.plot(x,x+4, color = 'black',linestyle = '--', marker = 'P', markerfacecolor="#FF0000")
ax.plot(x,x+6, color = 'black',linestyle = '--', marker = 'o')
```

##7)-Barplots

Barplots are important because they allows us to represent graphically the categorical data.

```python
country = ['india','japon','mexico','colombia','alemania']
population = [random.randint(500,2000) for i in range(1,len(country)+1)]
```

```python
plt.bar(country,population, width=0.5, color=['orange','blue','green','purple','teal'])
plt.xticks(np.arange(len(country)), ('India','Japon','Mexico','Colombia','Alemania'), rotation=45)
plt.title('Population by Country')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
plt.barh(country,population, color=['orange','blue','green','purple','teal'])
#plt.xticks(np.arange(len(country)), ('India','Japon','Mexico','Colombia','Alemania'))
plt.title('Population by Country')
plt.ylabel('Country')
plt.xlabel('Population')
plt.show()
```

##2.2-Distributions

```python
sns.set(style='ticks',palette='dark')
tips = sns.load_dataset('tips')
tips .head()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,4))
sns.histplot(data=tips,x='tip',
             bins=15,hue='sex', 
             stat='probability',
             ax=axes[0,0])
sns.histplot(data=tips,x='tip',
             bins=15,hue='sex',
             cumulative=True,
             stat='probability',
             ax=axes[0,1])
sns.histplot(data=tips,x='total_bill',
             bins=15,hue='sex',
             stat='percent',
             ax=axes[1,0])
sns.histplot(data=tips,x='total_bill',
             bins=15,hue='sex',
             cumulative=True,
             stat='percent',
             ax=axes[1,1])
             
plt.subplots_adjust(wspace=0.4)
plt.subplots_adjust(hspace=0.4)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,4))
sns.kdeplot(data=tips,x='tip',
            hue='sex',bw_adjust=True,
            ax=axes[0,0])
sns.kdeplot(data=tips,x='tip',
            hue='sex',cumulative=True,bw_adjust=True,
            ax=axes[0,1])
sns.kdeplot(data=tips,x='tip',
            hue='sex',shade=True,bw_adjust=True,
            ax=axes[1,0])
sns.kdeplot(data=tips,x='tip',
            hue='sex',cumulative=True,bw_adjust=True,
            shade=True,
            ax=axes[1,1])
plt.subplots_adjust(wspace=0.4)
plt.subplots_adjust(hspace=0.4)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python

sns.displot(data=tips,x='tip',hue='sex',stat='count',multiple='stack',kde=True)
plt.figure(figsize=(4,3))
plt.show()
```

```python
sns.set(style="dark", palette="muted")
sns.color_palette("flare", as_cmap=True)

fig, axes = plt.subplots(figsize=(4,3))
sns.kdeplot(data=tips, x='tip', hue='day', shade=True)
fig.suptitle('Density Plot',horizontalalignment='center')

plt.grid()
fig.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.ylabel('Total bill')
plt.xlabel('Day')
plt.title('Tips as a function of the day and sex')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.ylabel('Total bill')
plt.xlabel('Day')
plt.title('Tips as a function of the day and sex')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.ylabel('Total bill')
plt.xlabel('Day')
plt.title('Tips as a function of the day and sex')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color='0',marker='<')
plt.ylabel('Total bill')
plt.xlabel('Day')
plt.title('Tips as a function of the day and sex')
plt.show()
```

```python
figure, axes = plt.subplots(nrows=1,ncols=2,figsize=(8,4))
sns.violinplot(data=tips,x='day',y='total_bill',
               hue='sex',dodge=True,ax=axes[0])
sns.swarmplot(data=tips,x='day',y='total_bill',
              hue='sex',dodge=True,color='0',marker='<',
              ax=axes[0])
sns.violinplot(data=tips,x='day',y='total_bill',
               hue='sex',dodge=False,split=True,
               ax=axes[1])

plt.subplots_adjust(wspace=0.4)
axes[0].set_ylabel('Total bill')
axes[1].set_ylabel('Total bill')
axes[0].set_xlabel('Day')
axes[1].set_xlabel('Day')
axes[0].set_title('Tips as a function of the day and sex')
axes[1].set_title('Tips as a function of the day and sex')
plt.show()
```
