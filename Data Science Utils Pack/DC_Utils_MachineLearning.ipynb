KNN algorithm inspector

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def KNN_k_inspection(x_data, y_data, t_size, n_tests=13):
    X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=t_size, random_state=4)
    accuracy_scores = {"Number_of_neighbors": [], "Train_Accuracy": [], "Test_Accuracy": []}

    for i in range(1, n_tests + 1):
        clf = KNeighborsClassifier(n_neighbors=i)
        clf.fit(X_train, y_train)

        y_predicted = clf.predict(X_test)
        y_predicted_train = clf.predict(X_train)
        acc_knn = accuracy_score(y_test, y_predicted)
        acc_knn_train = accuracy_score(y_train, y_predicted_train)

        accuracy_scores["Number_of_neighbors"].append(i)
        accuracy_scores["Train_Accuracy"].append(acc_knn_train)
        accuracy_scores["Test_Accuracy"].append(acc_knn)

    return pd.DataFrame(accuracy_scores)
```

Metric tester

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def testers(y_pred, y_real):
  metric_func = [mean_squared_error, mean_absolute_error, r2_score]
  scores = {"sqrt_mse": [], "mse": [], "mean_abs_err": [], "R2": []}

  for func in metric_func:
    
    if func == mean_squared_error:
      mse = func(y_real, y_pred)
      scores["mse"].append(mse)
      scores["sqrt_mse"].append(np.sqrt(mse))
    elif func == mean_absolute_error:
      mae = func(y_real, y_pred)
      scores["mean_abs_err"].append(mae)
    elif func == r2_score:
      r2 = func(y_real, y_pred)
      scores["R2"].append(r2)

  return pd.DataFrame(scores)
```
